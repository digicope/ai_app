```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### 선그래프

```python
a = np.array([10,14,19,20,25])
print(a)
plt.plot(a)

a2 = np.arange(20)
print(a2)
plt.plot(a2)

plt.show()  # 새 프레임창에 그래프를 출력
```

```python
# 1차 함수(방정식)의 그래프 : f(x) = a*x + b, 직선의 방정식
x = np.arange(-4.5,5.0,0.5)
y = 2*x + 0
print(x,y)

plt.plot(x,y)
plt.show()
```

```python
# 2차 함수(방정식)의 그래프 : f(x) = a*x^2 + b, 포물선의 방정식
x = np.arange(-4.5,5.0,0.5)
y = 2*x**2 + 0
print(x,y)

plt.plot(x,y)    # 포물선
plt.plot(x,4*x)  # 직선
plt.plot(x,-3*x) # 직선
plt.show()
```

### 로그 플롯(log plot)

```python
# 상용로그 : 10
print(np.log10(100)) # 2.0

# 자연로그 :  e = 2.71828
print(np.log(100))                # log(100) ==> 4.605170185988092
print(np.exp(4.605170185988092))  # e**4.605170185988092 ==> 100.0
```

```python
df = pd.read_csv('transcount.csv')  
df = df.groupby('year').agg('mean')
df     # 연도별 CPU 판매량
```

```python
years = df.index.values            # x값
counts = df['trans_count'].values  # y값

print(type(years),type(counts))
print(years,counts)

plt.plot(years,counts) # 원본 데이터로 그래프를 출력
                       # 올바른 정보(변화량)를 알 수가 없다
                       # 1970~2000년 사이의 변화량이 매우 작음
plt.show()
```

```python
# y값을 log변환하여 그래프로 출력
plt.plot(years,np.log(counts))  # 자연로그 사용 로그변환
                                # 1970~2000년 사이의 변화량이 분명해짐
# plt.plot(years,np.log10(counts))   # 상용 로그 
plt.show()
```

```python
print(np.log1p(0))  # 1을 더한 후 log변환, log(0)이 -inf(무한대)가 되지 않도록

print(np.log(1))   # 0.0
print(np.log(0))   # -inf
# np.log1p(x) ==> np.log(x + 1)
```

```python
# 로그 플롯 : semilogy(),semilogx(),loglog()
years = df.index.values            # x값
counts = df['trans_count'].values  # y값

plt.semilogy(years,counts,'o') # semilogy함수는 y값을 로그변환하여 그래프로 출력

poly = np.polyfit(years,np.log(counts),deg=1)  # deg : 1차항, x가 1개 ==> 기울기도 1개
print(poly) # [ 3.61559210e-01 -7.05783195e+02] ,회귀계수
# H(x) = 3.61559210e-01*x - 7.05783195e+02 , 예측방정식(함수)

plt.semilogy(years,np.exp(np.polyval(poly,years))) # 예측함수의 직선
# plt.semilogy(years,np.exp(3.61559210e-01*years - 7.05783195e+02 ))
plt.show()

np.exp(np.polyval(poly,years)) # polyfit()으로 얻은 poly내에 기울기와
# y절편을 이용하여 예측함수에 x값(years)을 넣어 예측값(y,counts,H(x))을 구한다
```

### 분산 플롯(scatter plot) : scatter()함수, 버블차트

```python
df = pd.read_csv('transcount.csv')  
df = df.groupby('year').agg('mean')

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg('mean')


df = pd.merge(df,gpu,how='outer',left_index=True,right_index=True)
df = df.replace(np.nan,0)  # df.fillna(0)
df
```

```python
years = df.index.values  # x값
counts = df['trans_count'].values
cnt_log = np.log(counts) # 로그변환한 y값
gpu_counts = df['gpu_trans_count'].values # 버블의 크기

# plt.figure(figsize=(12,10)) # 그래프 사이즈의 크기 설정
plt.scatter(years,cnt_log, c = 200*years,
           s = 20 + 200*gpu_counts/gpu_counts.max(),
           alpha = 0.5) # c : 컬러, s : 버블의 크기, alpah : 투명도
plt.grid(True)
plt.xlabel('years')
plt.ylabel('trans_count')
plt.title('trans & gpu counts per years')
plt.show()
```

```python
# 그래프에 한글 출력 설정(Windows)
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # '맑은 고딕'으로 설정 
matplotlib.rcParams['axes.unicode_minus'] = False      # minus 숫자 표시
matplotlib.rcParams['axes.titlesize'] = 18             # title 의 사이즈 변경
plt.title('연간판매량')
plt.show()
```

```python
# 범례(legend), 주석(annotation), 예측 함수의 직선 그래프를 함께 출력
years = df.index.values  # x값
counts = df['trans_count'].values
cnt_log = np.log(counts) # 로그변환한 y값
gpu_counts = df['gpu_trans_count'].values # 버블의 크기

# 추가 코드

# plt.figure(figsize=(12,10)) # 그래프 사이즈의 크기 설정

# 예측 함수의 직선 그래프를 출력
poly = np.polyfit(years,cnt_log,deg=1)
plt.plot(years,np.polyval(poly,years),label='Poly Fit')

# annotaion(주석 달기)
gpu_start = gpu.index.values.min()        # x축 : 1997
y_ann = np.log(df.at[1997,'trans_count']) # y축 :15.913528485217045
ann_str = 'First GPU\n%d'%gpu_start
# print(y_ann)
# print(ann_str)
plt.annotate(ann_str,xy=(gpu_start,y_ann),
            arrowprops=dict(arrowstyle='->'),
            xytext=(-22,+35), textcoords='offset points')

# scatter plot
plt.scatter(years,cnt_log, c = 200*years,
           s = 20 + 200*gpu_counts/gpu_counts.max(),
           alpha = 0.5, label = 'Scatter Plot') # c : 컬러, s : 버블의 크기, alpah : 투명도

plt.grid(True)
plt.xlabel('years')
plt.ylabel('trans_count')
plt.title('trans & gpu counts per years', fontdict={'fontsize':18,'color':'b'})
plt.legend(loc='upper left')
plt.show()
```

```python
# 지연 플롯(lag plot) : time series data(시계열 테이터) 분석
# 시계열 데이터에서 x축과 y축의 값을 지연(lag) 값으로 차이나게 해서
# 시간의 변화에 따른 데이터의 변화 관계를 파악해보는 그래프

from pandas.plotting import lag_plot

# 지연값 : 1년,  lag = 1
# x축 : 기준 연도의 trans_count의 로그 값
# y축 : (기준 연도 + 1)의 trans_count의 로그 값
lag_plot(np.log(df['trans_count']),lag=1)
plt.grid()
plt.show()
```

```python
# 자기 상관 플롯
# 시계열의 시차 값(lagged values) 사이의 상관 관계를 출력
from pandas.plotting import autocorrelation_plot

print(df.corr())

# x축 : 시간 지연(lag), 1년 단위
# y축 : 상관 계수
autocorrelation_plot(np.log(df['trans_count']))
plt.grid(True)
plt.show()

# 지연값(lag)이 작으면 상관계수는 크다
# 지연값(lag)이 크면 상관계수는 작아지고, 0으로 수렴한다
```

### Boxplot(상자 그림)
* pandas
* matplotlib
* seaborn

```python
# pandas
df = pd.read_csv('transcount.csv')  
df = df.groupby('year').agg('mean')

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg('mean')

df = pd.merge(df,gpu,how='outer',left_index=True,right_index=True)

# 로그변환, 결측치는 0으로 
df['trans_count'] = np.log(df['trans_count'].values)
df['gpu_trans_count'] = np.log(df['gpu_trans_count'].values)

df = df.replace(np.nan,0)  # df.fillna(0)

print(df['trans_count'].median())  # 중위수, 15.520258650202699
print(df['trans_count'].mean())    # 평균, 15.283133541390544

df.plot.box()    # 결측치가 있어도 출력
plt.text(1,df['trans_count'].median(),round(df['trans_count'].median(),2))
plt.text(2,df['gpu_trans_count'].median(),round(df['gpu_trans_count'].median(),2))
plt.show()
```

```python
# matplotlib  : 결측치가 있을 경우 출력 안됨
df = df.replace(np.nan,0)  # df.fillna(0)
plt.boxplot([df['trans_count'],df['gpu_trans_count']])
plt.text(1,df['trans_count'].median(),round(df['trans_count'].median(),2))
plt.text(2,df['gpu_trans_count'].median(),round(df['gpu_trans_count'].median(),2))

plt.show()
```

```python
# seaborn  :  결측치가 있어도 출력
import seaborn as sns
sns.boxplot(data=df, order=['trans_count','gpu_trans_count'],
           width=0.15, linewidth=1.2)
plt.text(0,df['trans_count'].median(),round(df['trans_count'].median(),2))
plt.text(1,df['gpu_trans_count'].median(),round(df['gpu_trans_count'].median(),2))

plt.show()
```
